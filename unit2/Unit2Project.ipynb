#### Majors and Distribution of Students
----
This section of the report group the data by general majors and put them in a list in a dictionary. Then, we proceed on analyzing the enrollment data for different majors and uses a 5 number summary to showcase the outliers of the data.

```python
major_sort = reviews.groupby('Major_category')['Major'].apply(list).to_dict()
print(major_sort)
print("Number of departments:", len(major_sort))
major_size = reviews.groupby('Major_category').size()
major_size
```

```python
major_student_total = reviews.groupby('Major_category')["Total"].sum().astype(int)
major_student_total
```

```python
major_student_total.plot(kind = 'bar', title = "Total Students per Major Category", ylabel = "Total Students", xlabel = "Major Category", rot = 45, figsize = (20,10))

```

```python
reviews.plot(kind='scatter', x='Major_code', y='Total', title="Total Students by Major Code", xlabel="Major Code", ylabel="Total Students", figsize=(20, 10))
```

```python
enroll_summary = reviews['Total'].describe()
enroll_summary
```

```python
reviews['Total'].plot.box(title = "Box Plot of Total Students", figsize = (20, 10))
print("The hollow circles represents outliers of the dataset")
```

#### Z-score
---
Calculating the z-score of enrollment at different majors

```python
mean = 39370.081395
std = 63483.491009

reviews['z-score'] = (reviews['Total'] - mean) / std
print(reviews[['Total', 'z-score']])
```

#### Distribution of Gender
---
This section of the report analyzes the distribution of gender in each major category. In addition, we convert the number of enrolled females and males into percentages out of 100% to showcase the preference of gender for different majors upon enrollment.

```python
gender_distribution = reviews.groupby('Major_category')[['Men', 'Women']].sum().astype(int)
print(gender_distribution)
```

```python
gender_distribution_percent = gender_distribution.div(gender_distribution.sum(axis = 1), axis = 0) * 100
print(gender_distribution_percent)

gender_distribution_percent.plot(kind = 'bar', stacked = True, title = "Gender Distribution per Major Category", ylabel = "Number of Students", xlabel = "Major Category", rot = 45, figsize=(20, 10))
```

#### The Extremes of the Dataset
---
This section compares and isolates the highest enrolled major with the lowest enrolled major.

```python
highest_enroll = reviews.loc[reviews['Total'].idxmax()]
print(highest_enroll)
print()
print("-----------------------------------------")
print()
lowest_enroll = reviews.loc[reviews['Total'].idxmin()]
print(lowest_enroll)
```

#### Gender Correlation
---
This section of the report illustrates the correlation between gender in total enrolled population

```python
corr_student = reviews[['Total', 'Men', 'Women']].corr()
corr_student
```
