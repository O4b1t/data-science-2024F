Analysis for majors and distribution of students in each major

```python
major_sort = reviews.groupby('Major_category')['Major'].apply(list).to_dict()
print(major_sort)
print("Number of departments:", len(major_sort))
major_size = reviews.groupby('Major_category').size()
major_size
```

```python
major_student_total = reviews.groupby('Major_category')["Total"].sum().astype(int)
major_student_total
```

```python
major_student_total.plot(kind = 'bar', title = "Total Students per Major Category", ylabel = "Total Students", xlabel = "Major Category", rot = 45, figsize = (20,10))
major_student_total.hist()
```

Analysis of distribution of gender in each major category

```python
gender_distribution = reviews.groupby('Major_category')[['Men', 'Women']].sum().astype(int)
gender_distribution
```
